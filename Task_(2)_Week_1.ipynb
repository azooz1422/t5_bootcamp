```python
import seaborn as sns
f=df[['Datetime', 'TrafficIndexLive', 'JamsCount', 'JamsDelay',
       'JamsLength', 'TrafficIndexWeekAgo', 'TravelTimeHistoric',
       'TravelTimeLive']].corr()

sns.heatmap(f,annot=True)
```
